# Import libraries

```python
import pandas as pd
import numpy as np 
from transformers import AutoTokenizer, AutoModel

import psycopg2
from pgvector.psycopg2 import register_vector
import torch
import ollama
```

# 20NG Import

```python
df_20NG                 = pd.read_csv('20260211_20NG_Dataset.csv')
df_20NG_from            = df_20NG['From']
df_20NG_organization    = df_20NG['Organization']
df_20NG_newsgroup       = df_20NG['Newsgroup']
df_20NG_body            = df_20NG['Body']

from_organization_newsgroup_body_20NG       = df_20NG_from.reset_index(drop=True)    + ' ' + df_20NG_organization.reset_index(drop=True)    + ' ' + df_20NG_newsgroup.reset_index(drop=True)        + ' ' + df_20NG_body.reset_index(drop=True)
```

```python
dataset             = "20NG"
addi_info_inc       = "from_organization_newsgroup_body"
text_data           = from_organization_newsgroup_body_20NG
```

```python
def calc_single_embedding(chosen_record, model_name, max_length_val): 

    tokeniser           = AutoTokenizer.from_pretrained(model_name, local_files_only = True)
    model               = AutoModel.from_pretrained(model_name, local_files_only = True)

    inputs = tokeniser(chosen_record, return_tensors='pt', truncation=True, max_length=max_length_val)

    with torch.no_grad():
        outputs = model(**inputs)    

    # extract the last hidden layer - its embeddings
    token_embeddings = outputs.last_hidden_state 
    document_embedding = token_embeddings.mean(dim=1).squeeze(0)  
    document_embedding = document_embedding.cpu().numpy()

    return document_embedding
```

# Qwen3

```python
# qwen3 tokeniser
qwen_tokeniser = AutoTokenizer.from_pretrained(
    "./models/qwen3", 
    local_files_only = True
)

# function to truncate the text, to be used when varying qwen3 context length variable 
def truncate_text(text, tokeniser, context_length): 
    tokens = tokeniser.encode(text, add_special_tokens = False)
    tokens = tokens[:context_length]
    truncated_text = tokeniser.decode(tokens)
    return truncated_text
```

## 1024

```python
def create_send_qwen3_embeddings_ML1024(record, addi_info_inc, model_name, dataset, max_length_data):

    response    = ollama.embed(model=model_name, input = record)
    embedding   = response["embeddings"][0] 

    # Connect to PostgreSQL
    conn = psycopg2.connect(
        dbname  = "embeddings_db",
        user    = "",
        host    = "localhost",
        port    = "5432"
    )

    register_vector(conn)

    # Insert into postgres table
    cursor = conn.cursor()
    cursor.execute(
        "INSERT INTO qwen3_from_organization_newsgroup_body_20NG.qwen3_from_organization_newsgroup_body_ML1024_20NG (dataset, embedding_model, embedding_max_length, additional_info_inc, embedding_data) VALUES (%s, %s, %s, %s, %s)",
        (dataset, model_name, max_length_data, addi_info_inc, embedding)
    )

    conn.commit()
    cursor.close()
    conn.close()
```

```python
# create the Qwen3 embeddings and add them to postico
def create_Qwen3_embeddings_ML1024(model_name, text_data, tokeniser, max_len): 

    for i in range(len(text_data)): 
        trnctd_tkns = truncate_text(text_data[i], tokeniser, max_len)
        create_send_qwen3_embeddings_ML1024(trnctd_tkns, addi_info_inc, model_name, "20NG", max_len)
```

```python
model_name      = "qwen3-embedding:8b"
max_length_data = 1024

create_Qwen3_embeddings_ML1024(model_name, text_data, qwen_tokeniser, max_length_data)
```

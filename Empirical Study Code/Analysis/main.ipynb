```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
df_20NG = pd.read_csv('twenty_ng_results.csv'); 

df_20NG['additional_info']   = df_20NG['additional_info'].astype('category')
df_20NG['embedding_model']   = df_20NG['embedding_model'].astype('category')
df_20NG['dr_method']         = df_20NG['dr_method'].astype('category')

print("Ss range:", df_20NG['Ss'].min(), df_20NG['Ss'].max())
```

```python
df_SV = pd.read_csv('sv_results.csv'); 

df_SV['additional_info']   = df_SV['additional_info'].astype('category')
df_SV['embedding_model']   = df_SV['embedding_model'].astype('category')
df_SV['dr_method']         = df_SV['dr_method'].astype('category')

print("Ss range:", df_SV['Ss'].min(), df_SV['Ss'].max())
```

```python
anova_inter_interaction_20NG = smf.ols(
    formula='Ss ~ C(additional_info) * C(embedding_model) * C(dr_method)', 
    data=df_20NG
).fit()

interaction_ANOVA_20NG = sm.stats.anova_lm(anova_inter_interaction_20NG, typ=2)
print(interaction_ANOVA_20NG)
interaction_ANOVA_20NG.to_csv("anova_interaction_results_20NG.csv")
```

```python
anova_inter_interaction_SV = smf.ols(
    formula='Ss ~ C(additional_info) * C(embedding_model) * C(dr_method)', 
    data=df_SV
).fit()

interaction_ANOVA_SV = sm.stats.anova_lm(anova_inter_interaction_SV, typ=2)
print(interaction_ANOVA_SV)
interaction_ANOVA_SV.to_csv("anova_interaction_results_SV.csv")
```

## Averages

```python
EMs                     = ["allenai/longformer-base-4096", "google/bigbird-roberta-base", "qwen3-embedding:8b", "sentence-transformers/all-MiniLM-L6-v2", "google-bert/bert-base-uncased", "intfloat/e5-small-v2"]
DRMs                    = ["PCA", "UMAP", "PaCMAP"]
additional_info_20NG    = ["body_only", "from_body", "from_newsgroup_body", "from_organization_body", "from_organization_newsgroup_body", "newsgroup_body", "organization_body", "organization_newsgroup_body", "subject_body", "subject_from_body", "subject_from_newsgroup_body", "subject_from_organization_body", "subject_from_organization_newsgroup_body", "subject_newsgroup_body", "subject_organization_body", "subject_organization_newsgroup_body"]
additional_info_SV      = ["age_comment", "cah_age_comment", "cah_comment", "comment_only", "institution_age_comment", "institution_cah_comment", "institution_comment", "question_age_comment", "question_cah_age_comment", "question_cah_comment", "question_comment", "question_institution_age_comment", "question_institution_cah_age_comment", "question_institution_cah_comment", "institution_cah_age_comment", "question_institution_comment"]
short_EMs               = ["sentence-transformers/all-MiniLM-L6-v2", "google-bert/bert-base-uncased", "intfloat/e5-small-v2"]
long_EMs                = ["allenai/longformer-base-4096", "google/bigbird-roberta-base", "qwen3-embedding:8b"]
```

```python
def calc_average_EM_DRM_val(df, EM, DRM): 

    # separate by EM and DRM
    em_results      = df.loc[df['embedding_model'] == EM]
    em_drm_results  = em_results[(em_results['dr_method'] == DRM)]

    # get only the Ss column
    ss_results      = em_drm_results['Ss']
    
    # calculate the average
    ss_average      = ss_results.mean()
    return ss_average
```

```python
for i in range(len(EMs)):
    print(" ")
    for j in range(len(DRMs)):
        average = calc_average_EM_DRM_val(df_20NG, EMs[i], DRMs[j])
        print(f"EM: {EMs[i]}, DRM: {DRMs[j]}, average: {average.round(4)}")
```

```python
for i in range(len(EMs)):
    print(" ")
    for j in range(len(DRMs)):
        average = calc_average_EM_DRM_val(df_SV, EMs[i], DRMs[j])
        print(f"EM: {EMs[i]}, DRM: {DRMs[j]}, average: {average.round(4)}")
```

```python
def calc_average_additional_info_val(df, additional_info): 

    # separate by EM and DRM
    em_results      = df.loc[df['additional_info'] == additional_info]
  
    # get only the Ss column
    ss_results      = em_results['Ss']
    
    # calculate the average
    ss_average      = ss_results.mean()
    return ss_average
```

```python
def calc_average_shorter_EM_max_len(df, EM, DRM, metric): 

    # separate by EM and DRM
    em_results          = df.loc[df['embedding_model'] == EM]
    em_drm_results      = em_results[(em_results['dr_method'] == DRM)]

    em_drm_5dim_results     = em_drm_results[(em_drm_results['embedding_max_length'] == 1024)]
    em_drm_10dim_results    = em_drm_results[(em_drm_results['embedding_max_length'] == 2048)]
    em_drm_50dim_results    = em_drm_results[(em_drm_results['embedding_max_length'] == 4096)]

    # get only the Ss column
    ss_5dim_results         = em_drm_5dim_results[metric]
    ss_10dim_results        = em_drm_10dim_results[metric]
    ss_50dim_results        = em_drm_50dim_results[metric]
    
    # calculate the average
    ss_5dim_average         = ss_5dim_results.mean()
    ss_10dim_average        = ss_10dim_results.mean()
    ss_50dim_average        = ss_50dim_results.mean()

    return ss_5dim_average, ss_10dim_average, ss_50dim_average
```

```python
def calc_average_shorter_EM_DRM_ML_val(df, EM, DRM): 

    # separate by EM and DRM
    em_results          = df.loc[df['embedding_model'] == EM]
    em_drm_results          = em_results[(em_results['dr_method'] == DRM)]

    result_1            = em_drm_results[(em_drm_results['embedding_max_length'] == 128)]
    result_2            = em_drm_results[(em_drm_results['embedding_max_length'] == 256)]
    result_3            = em_drm_results[(em_drm_results['embedding_max_length'] == 512)]

    # get only the Ss column
    result_1_ss         = result_1['Ss']
    result_2_ss         = result_2['Ss']
    result_3_ss         = result_3['Ss']
    
    # calculate the average
    aver_result_1_ss    = result_1_ss.mean()
    aver_result_2_ss    = result_2_ss.mean()
    aver_result_3_ss    = result_3_ss.mean()

    return aver_result_1_ss, aver_result_2_ss, aver_result_3_ss
```

```python
def calc_average_longer_EM_DRM_ML_val(df, EM, DRM): 

    # separate by EM and DRM
    em_results          = df.loc[df['embedding_model'] == EM]
    em_drm_results          = em_results[(em_results['dr_method'] == DRM)]

    result_1            = em_drm_results[(em_drm_results['embedding_max_length'] == 1024)]
    result_2            = em_drm_results[(em_drm_results['embedding_max_length'] == 2048)]
    result_3            = em_drm_results[(em_drm_results['embedding_max_length'] == 4096)]

    # get only the Ss column
    result_1_ss         = result_1['Ss']
    result_2_ss         = result_2['Ss']
    result_3_ss         = result_3['Ss']
    
    # calculate the average
    aver_result_1_ss    = result_1_ss.mean()
    aver_result_2_ss    = result_2_ss.mean()
    aver_result_3_ss    = result_3_ss.mean()

    return aver_result_1_ss, aver_result_2_ss, aver_result_3_ss
```

```python
def calc_average_EM_DRM_dim_size_val(df, EM, DRM): 

    # separate by EM and DRM
    em_results              = df.loc[df['embedding_model'] == EM]
    em_drm_results          = em_results[(em_results['dr_method'] == DRM)]

    result_1                = em_drm_results[(em_drm_results['dimension_size'] == 5)]
    result_2                = em_drm_results[(em_drm_results['dimension_size'] == 10)]
    result_3                = em_drm_results[(em_drm_results['dimension_size'] == 50)]

    # get only the Ss column
    result_1_ss             = result_1['Ss']
    result_2_ss             = result_2['Ss']
    result_3_ss             = result_3['Ss']
    
    # calculate the average
    aver_result_1_ss        = result_1_ss.mean()
    aver_result_2_ss        = result_2_ss.mean()
    aver_result_3_ss        = result_3_ss.mean()

    return aver_result_1_ss, aver_result_2_ss, aver_result_3_ss
```

```python
def calc_aver_val_shorter(df, EM, DRM): 
    for i in range(len(DRM)): 
        print(" ")
        for j in range(len(EM)): 
            ss_128, ss_256, ss_512 = calc_average_shorter_EM_DRM_ML_val(df, EM[j], DRM[i])
            print(f"EM: {EM[j]}, DRMs: {DRM[i]}, ss_aver_128: {ss_128.round(4)}, ss_aver_256: {ss_256.round(4)}, ss_aver_512: {ss_512.round(4)}")

def calc_aver_val_longer(df, EM, DRM): 
    for i in range(len(DRM)): 
        print(" ")
        for j in range(len(EM)): 
            ss_1024, ss_2048, ss_4096 = calc_average_longer_EM_DRM_ML_val(df, EM[j], DRM[i])
            print(f"EM: {EM[j]}, DRMs: {DRM[i]}, ss_aver_1024: {ss_1024.round(4)}, ss_aver_2048: {ss_2048.round(4)}, ss_aver_4096: {ss_4096.round(4)}")

def calc_aver_dim_size(df, EM, DRM): 
    for i in range(len(DRM)): 
        print(" ")
        for j in range(len(EM)): 
            ss_5dim, ss_10dim, ss_50dim = calc_average_EM_DRM_dim_size_val(df, EM[j], DRM[i])
            print(f"EM: {EM[j]}, DRMs: {DRM[i]}, ss_5dim: {ss_5dim.round(4)}, ss_10dim: {ss_10dim.round(4)}, ss_50dim: {ss_50dim.round(4)}")
```

```python
calc_aver_val_shorter(df_SV, short_EMs, DRMs)
calc_aver_val_longer(df_SV, long_EMs, DRMs)
```
